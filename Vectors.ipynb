```python
import numpy as np
import matplotlib.pyplot as plt
```

# Vector plotting

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale = 1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',scale = 1,color='b')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
plt.quiver(0,0,4,5,angles='xy',scale = 1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
def printVector(vectors):
  for idx,vec in enumerate(vectors):
    colors = ['g','r','b','y']
    plt.quiver(vec[0],vec[1],vec[2],vec[3],scale_units = 'xy',angles='xy',color=colors[idx],scale=1)
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

```python
vecs = [(0,0,3,4),(0,0,-3,5),(0,0,5,-2),(0,0,-4,-3)]
printVector(vecs)
```

# Vector Addition

```python
vecs = [np.asarray([0,0,3,4]),np.asarray([0,0,-3,5]),np.asarray([0,0,5,-2]),np.asarray([0,0,-4,-3])]

printVector([vecs[0],vecs[1],vecs[0]+vecs[1]])
```

```python
printVector([vecs[2],vecs[1],vecs[2]+vecs[1]])
```

```python
printVector([vecs[2],vecs[1],vecs[2]-vecs[1]])
```

# Vector Projection

```python
printVector([vecs[0],vecs[2]])
```

Projecting vec[0] onto vec[2]

```python

scalar_projection = np.dot(vecs[0],vecs[2])/(np.linalg.norm(vecs[2]))
vec_projection = scalar_projection* vecs[2]
unit_vec_projection = scalar_projection * vecs[2]/np.linalg.norm(vecs[2])
printVector( [vecs[0],vecs[2],vec_projection])
printVector([vecs[0],vecs[2],unit_vec_projection])
```
